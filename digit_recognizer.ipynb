```python
# use all dataset for training (we already have test.csv for testing)
data=df.T
X=data[1:col]
Y=data[0]
```

```python
def Forward_p(X,W1,W2,B1,B2):
    Z1 = W1.dot(X)+B1
    A1= ReLu(Z1)
    Z2 = W2.dot(A1)+B2
    A2=SoftMax(Z2)
    return Z1,A1,Z2,A2
```

```python
def SoftMax(Z):
    expZ=np.exp(Z-np.max(Z,axis=0))
    return expZ/np.sum(expZ,axis=0,keepdims=True)
```

```python
def Back_p(X,Y,Z1,Z2,A1,A2,W1,W2):
    m=X.shape[1]
    Y=One_Hot(Y).T
    dL_dZ2= A2 - Y
    dL_dW2 = 1/m * dL_dZ2.dot(A1.T)
    dL_dB2 = 1/m * np.sum(dL_dZ2,axis=1,keepdims=True)
    dL_dZ1 = (W2.T.dot(dL_dZ2)) * Derive_ReLu(Z1)
    dL_dW1 = 1/m * dL_dZ1.dot(X.T)
    dL_dB1 = 1/m * np.sum(dL_dZ1,axis=1,keepdims=True)
    return dL_dW1,dL_dW2,dL_dB1,dL_dB2
```

```python
#activation function => relu
def ReLu(Z):
    return np.maximum(0,Z)

def Derive_ReLu(Z):
    return Z>0
```

```python
def Grad_Desc(X,Y,learn_rate,itera):
    W1,W2,B1,B2=params()
    for i in range(itera):
        Z1,A1,Z2,A2 = Forward_p(X,W1,W2,B1,B2)
        dL_dW1,dL_dW2,dL_dB1,dL_dB2 = Back_p(X, Y, Z1, Z2, A1, A2, W1, W2)
        W1,W2,B1,B2 = Update(W1,W2,B1,B2,dL_dW1,dL_dW2,dL_dB1,dL_dB2,learn_rate)
        if i%100==0:
            print(f"iteration:{i}")
            pred=Pred_Max(A2)
            accuracy=acc(Y,pred)
            print(accuracy)
    return W1,W2,B1,B2
```

```python
def One_Hot(Y):
    return np.eye(Y.max()+1)[Y]
```

```python
# initialize random params
def params():
    np.random.seed(9)
    W1 = np.random.randn(64,784)*np.sqrt(1./784)
    W2 = np.random.randn(10,64)*np.sqrt(1./64)
    B1 = np.zeros((64,1))
    B2 = np.zeros((10,1))
    return W1,W2,B1,B2
```

```python
#update params
def Update(W1,W2,B1,B2,dL_dW1,dL_dW2,dL_dB1,dL_dB2,learn_rate):
    W1 = W1-learn_rate*dL_dW1
    W2 = W2-learn_rate*dL_dW2
    B1 = B1-learn_rate*dL_dB1
    B2 = B2-learn_rate*dL_dB2
    return W1,W2,B1,B2
```

```python
def Pred_Max(A2):
    return np.argmax(A2,0)
```

```python
def acc(Y,pred):
    return np.sum(Y==pred)/Y.size
```

```python
W1,W2,B1,B2 = Grad_Desc(X,Y, 0.01, 1200)
```
