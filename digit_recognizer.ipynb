```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("./train.csv")
df.head()
```

```python
row,col=df.shape
```

```python
df=np.array(df)
```

```python
# use all dataset for training (we already have test.csv for testing)
data=df.T
X=data[1:col]/255.0 
Y=data[0]
```

```python
def Forward_p(X,W1,W2,W3,B1,B2,B3):
    Z1 = W1.dot(X)+B1
    A1= ReLu(Z1)
    Z2 = W2.dot(A1)+B2
    A2=ReLu(Z2)
    Z3=W3.dot(A2)+B3
    A3=SoftMax(Z3)
    return Z1,A1,Z2,A2,Z3,A3
```

```python
def SoftMax(Z):
    expZ=np.exp(Z-np.max(Z,axis=0))
    return expZ/np.sum(expZ,axis=0,keepdims=True)
```

```python
def Back_p(X,Y,Z1,Z2,Z3,A1,A2,A3,W1,W2,W3):
    m=X.shape[1]
    Y=One_Hot(Y).T
    dL_dZ3 = A3 - Y
    dL_dW3 = 1/m * dL_dZ3.dot(A2.T)
    dL_dB3 = 1/m * np.sum(dL_dZ3, axis=1, keepdims=True)
    dL_dZ2 = W3.T.dot(dL_dZ3) * (Z2 > 0)
    dL_dW2 = 1/m * dL_dZ2.dot(A1.T)
    dL_dB2 = 1/m * np.sum(dL_dZ2, axis=1, keepdims=True)
    dL_dZ1 = W2.T.dot(dL_dZ2) * (Z1 > 0)
    dL_dW1 = 1/m * dL_dZ1.dot(X.T)
    dL_dB1 = 1/m * np.sum(dL_dZ1, axis=1, keepdims=True)
    return dL_dW1,dL_dW2,dL_dW3,dL_dB1,dL_dB2,dL_dB3
```

```python
#activation function => relu
def ReLu(Z):
    return np.maximum(0,Z)

def Derive_ReLu(Z):
    return Z>0
```

```python
def Grad_Desc(X,Y,learn_rate,itera):
    W1,W2,W3,B1,B2,B3=params()
    for i in range(itera):
        Z1,A1,Z2,A2,Z3,A3 = Forward_p(X,W1,W2,W3,B1,B2,B3)
        dL_dW1,dL_dW2,dL_dW3,dL_dB1,dL_dB2,dL_dB3 = Back_p(X, Y, Z1, Z2,Z3, A1, A2,A3, W1, W2,W3)
        W1,W2,W3,B1,B2,B3 = Update(W1,W2,W3,B1,B2,B3,dL_dW1,dL_dW2,dL_dW3,dL_dB1,dL_dB2,dL_dB3,learn_rate)
        if i%100==0:
            print(f"iteration:{i}")
            pred=Pred_Max(A3)
            accuracy=acc(Y,pred)
            print(accuracy)
    return W1,W2,W3,B1,B2,B3
```

```python
def One_Hot(Y):
    return np.eye(Y.max()+1)[Y]
```

```python
# initialize random params
def params():
    np.random.seed(9)
    W1 = np.random.randn(64,784)*np.sqrt(1./784)
    W2 = np.random.randn(8,64)*np.sqrt(1./64)
    W3 = np.random.randn(10,8)*np.sqrt(1./8)
    B1 = np.zeros((64,1))
    B2 = np.zeros((8,1))
    B3 = np.zeros((10,1))
    return W1,W2,W3,B1,B2,B3
```

```python
#update params
def Update(W1,W2,W3,B1,B2,B3,dL_dW1,dL_dW2,dL_dW3,dL_dB1,dL_dB2,dL_dB3,learn_rate):
    W1 = W1-learn_rate*dL_dW1
    W2 = W2-learn_rate*dL_dW2
    W3 = W3-learn_rate*dL_dW3
    B1 = B1-learn_rate*dL_dB1
    B2 = B2-learn_rate*dL_dB2
    B3 = B3-learn_rate*dL_dB3
    return W1,W2,W3,B1,B2,B3
```

```python
def Pred_Max(A3):
    return np.argmax(A3,0)
```

```python
def acc(Y,pred):
    return np.sum(Y==pred)/Y.size
```

```python
W1,W2,W3,B1,B2,B3 = Grad_Desc(X,Y, 0.1, 6000)
```

### testing

```python
_, _, _,_,_, A3_test = Forward_p(X_test,W1,W2,W3,B1,B2,B3)
```

```python
pred=Pred_Max(A3_test)
print(pred)
```
