```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('student_performance_dataset.csv')
```

```python
# for now we consider only two input params and one label
df = df[['final_exam_score', 'study_hours_per_day', 'sleep_hours']]
df = df.dropna()
row,col=df.shape
df=np.array(df)
split_idx=int(row * 0.2)
train_data=df[split_idx:row].T
X_train=train_data[1:col]
Y_train=train_data[0:1]
test_data=df[:split_idx].T
X_test=test_data[1:col]
Y_test=test_data[0:1]
```

```python
def forward_prop(X,W1,B1):
    Z1=W1.dot(X)+B1
    A1=Z1
    return Z1,A1
```

```python
# loss_function => mse(mean square error)
def loss_f(A1,Y):
    return np.mean((A1-Y)**2)
```

```python
def initilize_params():
    np.random.seed(42) 
    W1=np.random.randn(1,2)*np.sqrt(1./2)
    B1=np.zeros((1,1))
    return W1, B1
```

```python
def back_prop(X, Y, Z1, A1, W1):
    m=X.shape[1]
    dZ1=2*(A1-Y)
    dL_dW1=1/m*dZ1.dot(X.T)
    dL_dB1=1/m*np.sum(dZ1,axis=1,keepdims=True)
    return dL_dW1,dL_dB1
```

```python
def update(W1, B1, dL_dW1, dL_dB1, lrn_rate):
    W1=W1-lrn_rate*dL_dW1
    B1=B1-lrn_rate*dL_dB1
    return W1, B1
```

```python
def gradient_descent(X, Y, lrn_rate, itrs):
    W1,B1=initilize_params()
    for i in range(itrs):
        Z1, A1 = forward_prop(X, W1, B1)        
        dL_dW1, dL_dB1 = back_prop(X, Y, Z1, A1, W1)
        W1, B1 = update(W1, B1, dL_dW1, dL_dB1, lrn_rate)
        if i % 1000 == 0:
            loss = loss_f(A1, Y)
            print(f'Iteration: {i:4d} Loss: {loss:.2f}')
    return W1, B1
```

```python
final_W1, final_B1 = gradient_descent(X_train, Y_train, 0.01, 10000)
```

```python
def get_accuracy(predictions,Y):
    mape=np.mean(np.abs(predictions-Y) / Y)*100
    return 100 - mape
```

```python
predictions=forward_prop(X_test,final_W1,final_B1)
acc=get_accuracy(predictions,Y_test)
print(acc)
```
